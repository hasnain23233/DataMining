```python
x = data[['Pregnancies' , 	'Glucose' ,	'BloodPressure' ,	'SkinThickness' ,	'Insulin' ,	'BMI' , 	'DiabetesPedigreeFunction'	,'Age']]
x
```

```python
y = data['Outcome']
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 , random_state = 5)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 1000)
```

```python
model.fit(x_train , y_train)
```

```python
model.coef_
```

```python
pridictions = model.predict(x_test)
pridictions
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
cof_matix = confusion_matrix(y_test , pridictions)
plt.figure(dpi = 100)
sns.heatmap(cof_matix , annot = True , fmt = 'd' , cmap = 'Blues')
plt.xlabel('Pridicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
Result = pd.DataFrame({'Actual' : y_test , "Priduction" : pridictions})
Result.head(100)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test , pridictions)
score
```
