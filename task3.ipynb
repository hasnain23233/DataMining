```python
x = data[['fixed acidity' , 	'volatile acidity',	'citric acid'	,'residual sugar',	'chlorides',	'free sulfur dioxide' ,	'total sulfur dioxide' ,	'density'	,'pH'	,'sulphates' ,	'alcohol']]
x
```

```python
y = data['quality']
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 ,random_state = 42)
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 1000)
model.fit(x_train , y_train)
```

```python
priduction = model.predict(x_test)
priduction
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
cof_matix = confusion_matrix(y_test , priduction)
plt.figure(dpi = 100)
sns.heatmap(cof_matix , annot = True , fmt = 'd' , cmap = 'Blues')
plt.xlabel('Pridicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
result = pd.DataFrame({'Actual': y_test , 'Priduction': priduction , 'difference': y_test - priduction})
result
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test , priduction)
score
```
